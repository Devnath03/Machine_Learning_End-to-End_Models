```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pickle
```

```python
#Load The CSV DataFile
Startup = pd.read_csv('Startups_Dataset_Sample.csv')
```

```python
#Execute the Saple Datset
Startup
```

```python
#Check the Shape of the DataSet
Startup.shape
```

```python
#Check the Unique data
Startup.State.unique()
```

```python
#Check the Isnull of the DataSet
Startup.isnull().sum()
```

```python
#Convert String Data Into Int Data types
#Use the Label encoder
labelEncoder = LabelEncoder()
Startup['State'] = labelEncoder.fit_transform(Startup['State'])
```

```python
#Check the Converted Data
Startup.head()
```

```python
#Execute the data to array from dataset (Two Demention)
x = np.array(Startup.drop(columns=['Profit']))
y = np.array(Startup['Profit'])
```

```python
#Get the data to train the model
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Execute the Data for Train
x_train
```

```python
#Execute the Data for Test
x_test
```

```python
#Execute the Data for Train
y_train
```

```python
#Execute the Data for Test
y_test
```

```python
#Train the Model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#Create a sample prediction  get method
predict = model.predict([[168552,145823,412852,1]])
predict[0]
```

```python
# Check the Model Accuracy
Accuracy = model.score(x_test,y_test)
Accuracy
```

```python
#Check the coefficient the Dataset
model.coef_
```

```python
#Check coefficient from dataset
pd.DataFrame(model.coef_,["R&D_Spend","Administration","Marketing_Spend","State	Profit"],columns=['coef'])
```

```python
#Save the Model
with open("model.pickel","wb") as file:
    pickle.dump(model,file)
```
